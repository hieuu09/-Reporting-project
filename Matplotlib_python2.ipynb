```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
cities = pd.read_csv('https://raw.githubusercontent.com/CodexploreRepo/data-science/main/Code/A4_Matplotlib/data/california_cities.csv', sep =',')
```

```python
cities.head()
cities["area_total_km2"].sort_values(ascending=False);
# lat ci do cac dg gach nganh trai dat
# kinh do
```

```python
# latd vi do  , longd  kinh do
lat, lon = cities['latd'], cities['longd']# lay kinh do va vi do gan vao cac bien
```

```python

# latd vi do  , longd  kinh do
lat, lon = cities['latd'], cities['longd']# lay kinh do va vi do gan vao cac bien
population, area = cities['population_total'], cities['area_total_km2']
#kich thuoc
plt.figure(figsize=(8, 6));
# thay doi stle
plt.style.use('seaborn');
plt.scatter(lon, lat, c=np.log10(population), cmap='viridis', s= area, linewidths=0, alpha=0.5);
plt.axis('equal'); # kinh do va vi do bang nhau
plt.xlabel('Longtitule');
plt.ylabel('Latitude');
# do mau nam khoang
plt.colorbar(label='log$_10$(population)');
plt.clim(3, 7); # phan lhoang mau chi hien 3 den 7
# tao vung khaong mau dan sao

area_range =[50, 100, 300, 500]
for area in area_range:
  plt.scatter([], [], s=area, c='k', alpha=0.4,  label=str(area)+'km$^2$')

# scatterpoints voi moi 1 diem thi hien 1 diem
plt.legend(labelspacing =1, title= 'City Area', scatterpoints=1) # khoang cach cacs vong trong


# tao legend kich thuoc tung thanh pho 1
plt.title('California cities : population anf area distribution')
```
